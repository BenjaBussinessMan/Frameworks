```python
import asyncio
import nest_asyncio  # <--- 1. Importar
from semantic_kernel import Kernel
from semantic_kernel.connectors.ai.open_ai import OpenAIChatCompletion
from semantic_kernel.functions import KernelArguments
```

```python
nest_asyncio.apply()
```

```python
async def main():
    # 1. Inicializar el Kernel
    kernel = Kernel()

    # 2. Configurar el servicio de IA (OpenAI en este caso)
    service_id = "asistente_chat"
    kernel.add_service(
        OpenAIChatCompletion(
            service_id=service_id,
            api_key="TU_API_KEY_AQUI",
            ai_model_id="gpt-4o"
        )
    )

    # 3. Definir una "Plugin" semntica (Prompt Engineering)
    # En Semantic Kernel, los prompts se tratan como funciones.
    prompt_template = """
    Eres un experto en atencin al cliente. Analiza el siguiente comentario:
    
    Comentario: {{$input}}
    
    Por favor, responde con el siguiente formato:
    - Sentimiento: (Positivo/Negativo/Neutral)
    - Resumen: (Mximo 10 palabras)
    - Accin recomendada: (Breve sugerencia)
    """

    # Registrar la funcin en el kernel
    analisis_feedback = kernel.add_function(
        function_name="AnalizarFeedback",
        plugin_name="MarketingPlugin",
        prompt=prompt_template
    )

    # 4. Caso prctico: Ejecucin con un comentario real
    comentario_cliente = """
    El producto lleg dos das tarde y la caja estaba un poco abollada, 
    pero debo decir que el equipo funciona de maravilla y el soporte tcnico 
    fue muy amable al ayudarme con la configuracin inicial.
    """

    # Ejecutar la funcin
    resultado = await kernel.invoke(
        analisis_feedback, 
        KernelArguments(input=comentario_cliente)
    )

    print("--- ANLISIS DEL ASISTENTE ---")
    print(resultado)
```

```python
if __name__ == "__main__":
    asyncio.run(main()) # <--- Ahora esto ya no dar error
```
